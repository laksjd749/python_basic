```python
# [()]를 dict으로 변환
data = [
    ("apple", 1000),
    ("banana", 2000),
    ("orange", 3000)
]
#key 자리에는 불변 객체만 올 수 있다. unique

fruit = dict(data)

print(fruit)
```

```python
student = {"name": "홍길동"}
print(student)

# 요소조회
print(student["name"])

# 추가
student["age"] = 20
print(student)

# 수정
student["age"] = 30
print(student)

# 삭제
del student["age"]
print(student)

# in 연산
print("name" in student)
print("age" in student)
```

```python
student = {"name": "홍길동"}

print(student.get("name"))
print(student.get("age")) # 없으면 None 반환

student = {"name": "홍길동", "age": 20}
print(student.keys())
print(student.values())
print(student.items())

student.update({"age": 30, "score": 90})
print(student)

age = student.pop("age")
print(age)
print(student)

student.clear()
print(student)
```

```python
student = {
    "name": "홍길동",
    "age": 20
}
# 키
for key in student:
    print(key)
print("===========================")

# 값
for value in student.values():
    print(value)
print("===========================")

# 키 + 값   -> unpacking
for key, *value in student.items():
    print(key, value)
```
