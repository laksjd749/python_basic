```python
# 빈 딕셔너리
d1 = {}
d2 = dict()

print(d1)
print(d2)
```

```python
# {}사용
student = {
    "name" : "홍길동",
    "age": 20,
    "score": 90
}
print(student)
print(type(student))
```

```python
# dict() 사용
person = dict(name="김철수", age=30, city="서울")
print(person)
```

```python
# [()]를 dict으로 변환
data = [
    ("apple", 1000),
    ("banana", 2000),
    ("orange", 3000)
]
#key 자리에는 불변 객체만 올 수 있다. unique

fruit = dict(data)

print(fruit)
```

```python
d = {"name": "홍길동"}
print(d)

d = {100: "학생"}
print(d)

d = {
    (1, 2): "좌표"
}
print(d)

"""
# error
d = {
    [1, 2]: "리스트"
}
print(d)
"""
```

```python
student = {"name": "홍길동"}
print(student)

# 요소조회
print(student["name"])

# 추가
student["age"] = 20
print(student)

# 수정
student["age"] = 30
print(student)

# 삭제
del student["age"]
print(student)

# in 연산
print("name" in student)
print("age" in student)
```
