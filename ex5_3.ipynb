```python
nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # [20, 30, 40]
print(nums[:3]) # [10, 20, 30]
print(nums[2:]) # [30, 40, 50]
print(nums) #  [10, 20, 30, 40, 50] 슬라이싱 연산 후에도 리스트 자신은 변경되지 않는다.
```

```python
# 2칸씩 건너뛰기
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[::2])

nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[1::2])
```

```python
# 역순 출력
nums = [10, 20, 30, 40, 50]
print(nums[::-1]) # [50, 40, 30, 20, 10]

# 뒤에서 3개 추출
nums = [10, 20, 30, 40, 50, 60]
print(nums[-3:])

# 뒤에서 4개 중 2칸씩
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-4::2])

# 뒤에서 앞으로 2칸씩
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[::-2])
```

```python
# [5]위치부터 뒤로(-1) [1]위치 전까지
nums = [10, 20, 30, 40, 50, 60, 70]
print(nums[5:1:-1]) # 5위치부터 1전까지 -1칸식  [60, 50, 40, 30]

# 마지막 5개를 역순으로
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-1:-6:-1])
```

```python
# 슬라이싱 할당(슬라이싱 구간에 대입문)
nums = [10, 20, 30, 40, 50]
nums[1:4] = [7, 8] # 1~3구간 [20, 30, 40]에 [7, 8]을 대입 
print(nums) # [10, 7, 8, 50]

# 슬라이싱 결과를 다른 변수에 대입
a = [10, 20, 30, 40, 50, 60]
b = a[:]
c = a[::-1]
d = a[1::2]
e = a[:2]

print(b)
print(c)
print(d)
print(e)
```

```python
# append() : 맨 뒤에 1개 추가
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)
```

```python
# insert() : 원하는 위치에 추가
fruits = ["apple", "banana"]
fruits.insert(0, "orange")
print(fruits)
```

```python
# extend() : 여러 개 추가
nums = [1, 2, 3]
nums.extend([4, 5, 6])
print(nums)
```

```python
# remove() : 값으로 삭제
fruits = ["apple", "banana", "orange"]
fruits.remove("banana")
print(fruits)

# 수정은 인덱스 할당, 슬라이싱 할당으로 수정
```

```python
# pop() : 인덱스로 삭제 후 반환
nums = [10, 20, 30]
result = nums.pop() # 마지막 인덱스 삭제 후 반환
print(result)
print(nums)

nums = [10, 20, 30]
result = nums.pop(1) # 인덱스1 삭제 후 반환
print(result)
print(nums)
```

```python
# clear() : 전체 삭제
nums = [10, 20, 30]
nums.clear()
print(nums)
```

```python
# 정렬 메서드
nums = [50, 10, 40, 20]
n2 = nums.sort() # 오름차순     식임. 식은 반환값을 가진다.
print(nums)
print(n2)

nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 내림차순
print(nums)
```

```python
nums = [50, 10, 40, 20]
result = sorted(nums)
print(nums)
print(result)
```

```python
nums = [50, 10, 40, 20]
result = sorted(nums, reverse=True)
print(result)
```
