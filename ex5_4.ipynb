```python
a={1,2,3}
b={3,4,5}

#교집합
print(a&b)
print(a.intersection(b))

#합집합
print(a|b)
print(a.union(b))

#차집합
print(a-b)
print(a.difference(b))
```

```python
s={1,2,3}

s.add(4)    #하나 추가
print(s)

s.update([5,6]) #여러개 추가
print(s)

s.remove(2)     #값 2 삭제 (없으면 오류 발생)
print(s)

s.discard(10)   #값 없으면 무시
```

```python
s={1,2,3}

print(len(s))   #개수
print(2 in s)   #포함여부

s.clear()   #전체 삭제
print(s)    #set()
```

```python
fs = frozenset([1,2,3])
print(fs)
#fs.add(4)   #오류. 읽기 전용 set은 수정 불가
```

```python
# 패킹, 언패킹시 *연산자
# a,b=[1,2,3,4]     too much values to unpack
*a,b = [1,2,3,4]

print(a)
print(b)

print("============")

c,*d = (1,2,3,4)
print(c)
print(d)

print("============")

e,*f,g=(1,2,3,4)
print(e)
print(f)
print(g)
```
