```python
# 1. 정수 (int)
a = 10
print(a)
print(type(a))
```

```python
#2. 부동소수점 수(float) : 오차를 가진다
b = 3.14
print(b)
print(type(b))

print(0.1+0.2)
```

```python
# 3. 복소수 (complex)
c = 3 + 4j
print(c)
print(type(c))
```

```python
# 4. 논리값 (bool)
d = True
print(d)
print(type(d))
```

```python
# 5. 문자열 (str)
e = "Python"
print(e)
print(type(e))

print(e[1])
#e[1] = "K" -> error:문자열 객체는 불변.
```

```python
# 6. 리스트 (list)  <-가변
f = [1, 2, 3]
print(f)
print(type(f))

print(f[1])

f[1] = "abc"
print(f)
```

```python
# 7. 튜플 (tuple)   <- 불변
g = (1, 2, 3)
print(g)
print(type(g))

print(g[1])

# g[1] = 9 -> error: 튜플은 불변. 수정불가
```

```python
# 8. 딕셔너리 (dict)    <- 가변
h = {"id": "admin", "pw" : "pw1234"}
g= {(1,2,5,7):"pw1234"}

print(h)
print(g)

print(type(h))

print(h["id"])

h["pw"] = "key1234"
print(h)
```

```python
# 9. 집합 (set)
i = {1, 2, 3}
l = {1,2,3,2}

print(i)
print(l)
print(type(i))

#print(i[0])
```

```python
# 10. 범위 (range)
j = range(5)    #range(0,5)와 동일
print(j)
print(type(j))

print(list(j))
```

```python
# 14. 함수 (function)
def hello():
    print("Hello")
    print("bye")

print("hi")

hello()

print(hello)    #hello도 하나의 변수. 
print(type(hello))
```
